```python
このアルゴリズムについて
最初Aに任意の数字を与える。この数字を初期値と呼ぶ。Aがセーフとなり、次の数字を探すとき、A*=2Aとする。
Aが任意の数字Bを超えたとき、A+=1とする。
例えば、初期値が4000、B=10000とすると、A1=4000,A2=8000,A3=16000,A4=16001,A5=16002,   
この数当てゲームでは、正解するより、Aが大き数字でセーフを取り続けたほうが総合得点は増えるため。
以下のシュミレーションでは、一回のゲームでの回答回数を10000回にし、合計500回ゲームしたときの総合得点の平均を初期値を変化させながら追った。
シュミレーションのため、Qはランダムである。

考察
初期値が極端に大きくない限り、その大きさはあまり結果に影響しない。
それよりも、初期値とBの関係の方が重要。
例えば、A=9000,B=10000のとき、A=18000となり、ほぼほぼアウト。
つまり、B=10000のとき、初期値、もしくはAが増えていき、Bの少し下の数字になると、総合得点は大きく減る。
B=7500の場合のようにBがはあまり関係ない。
A,Bの組み合わせで最も良い組み合わせは何かを求める必要があるが、アウトじゃない限り、あまり差がないようにも感じる。

今後の課題
このアルゴリズムはQが大きい場合に強いが、Qが小さい数字に固まっている場合の対策が必要。
案としてはアウトの数を数えて、ある一定の基準を超えたら違うアルゴリズムを起動させるようにするなど。
```

```python
#Aを100からスタートさせ、#200刻みに20000まで変化させた。B=10000

import random
import numpy as np
import matplotlib.pyplot as plt

trials = 10000
ex = 500

def main():
    scores = []
    A_values = np.arange(100, 20000, 200)
   
    for A in A_values:
        total_score = 0
        
        for i in range(ex):
            score = 0.0  # floatに変更
            attempts = 0
            out = 0
            
            while attempts < trials:
                Q = random.randint(1, 20000)
                temp_A = A
                
                while True:
                    if temp_A == Q:
                        score += temp_A
                        break
                    elif temp_A > Q:
                        score -= (temp_A - Q) * 200
                        out += 1
                        break
                    else:
                        score += temp_A
                    
                    attempts += 1
                    
                    if temp_A > 10000:
                        temp_A += 1
                    else:
                        temp_A *= 2
            
            total_score += score / ex
        
        scores.append(total_score)
    
    plt.plot(A_values, scores)
    plt.xlabel("A")
    plt.ylabel("The average of total scores")
    plt.show()

main()
```
